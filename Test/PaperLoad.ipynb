```python
from dotenv import load_dotenv
import os
load_dotenv()
marker_api = os.getenv("MARKER")
filepath = '../TestPaper/基于跨地域混合云架构的人工智能模型推理平台.pdf'
```

# Marker API

```python
import requests

url = "https://www.datalab.to/api/v1/marker"

form_data = {
    'file': ('test.pdf', open('../TestPaper/基于跨地域混合云架构的人工智能模型推理平台.pdf', 'rb'), 'application/pdf'),
    'langs': (None, "English"),
    "force_ocr": (None, False),
    "paginate": (None, False),
    'output_format': (None, 'markdown'),
    "use_llm": (None, False),
    "strip_existing_ocr": (None, False),
    "disable_image_extraction": (None, False)
}

headers = {"X-Api-Key": marker_api}

response = requests.post(url, files=form_data, headers=headers)
data = response.json()
```

```python
import time

max_polls = 300
check_url = data["request_check_url"]

for i in range(max_polls):
    time.sleep(2)
    response = requests.get(check_url, headers=headers) # Don't forget to send the auth headers
    data = response.json()
    print(data)

    if data["status"] == "complete":
        break
```
